### Un-supervised Learning

Unsupervised learning is a type of machine learning that deals with previously undetected patterns in a data with no labels provided and with a minimum of human supervision. In contrast to supervised learning techniques which usually makes use of labeled data, unsupervised learning, allows modeling of probability densities over inputs.

Clustering is an example of un-supervised learning. We will be looking at one of the most basic clustering algorithm called K-mean clustering in this tutorial.

#### What is Clustering?

In machine learning, clustering is an example of separating the data points into specific groups in a way that similar data points are assigned to same groups. These data points of one group are more similar to the data points in same group and dissimilar to data points in other groups. Clustering is basically collection of data based on similarity and dissimilarity among them.
For example. Grouping of flowers species can be an example of clustering. Which we'll later see in this tutorial.

### K-mean Clustering

The goal of K-mean clustering algorithm is to divide the data into groups such that the total sum of squared distances from each point to the mean point of the cluster is minimized. We will be using Euclidean distance as a measurement to compute similarity between data. 

#### Advantages and Disadvantages of K-mean

##### Advantages
   - Very simple to implement.
   - Very fast.
   - Algorithm easily adapts to new examples.
   - Convergence is guaranteed
   - Scales to large dataset.
    
##### Disadvantages
   - Deciding the value of k manually is very difficult.
   - Outliers are problem.
   - Is dependent on initial values.
   - Problem when clustering data is of varying sizes and density.

### Algorithm of K-mean Clustering

K-mean algortihm works in an iterative process which is.

   - Pick some value of k i.e. number of clusters to create.
   - Initialize k "centroids" i.e. starting points in your data.
   - Create the clusters. Assign each point to the nearest centroid of the cluster.
   - Try to improve the clusters. Move each centroid to the center of its cluster.
   - Repeat above two steps until your centroids converge.
   - End Algorithm.

### Implementation using python

These are some of the essential libraries that are needed to use some built-in functionalities provided by python. Some well-known libraries are like pandas, matplotlib, sklearn.

Sklearn provide a lot of support for machine learning algorithms.

```python
from sklearn import datasets
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans
```

#### Dataset

For this tutorial we will be using IRIS dataset provided in sklearn library. This data consists of 50 samples from each of three species of Iris (Iris setosa, Iris virginica and Iris versicolor). It has four features from each sample: length and width of sepals and petals.

##### Loading dataset

```python
# Loading data
iris = datasets.load_iris()
```

Defining target and predictors

```python
X = iris.data[:, :2]
y = iris.target
```

Let's take a look at our data how it is distributed. We will use matplotlib's scatter plot to visualize out data. 

```python
plt.scatter(X[:,0], X[:,1], c=y, cmap='gist_rainbow')
plt.xlabel('Sepa1 Length')
plt.ylabel('Sepal Width')
```

Now, let's use k-mean algorithm provided by the sklearn library to fit data. As we can see in the above plot that there are three clusters so we will use k = 3

```python
# Create Kmean object using KMeans()
kmean = KMeans(n_clusters = 3, random_state=1)
# Fit on data
kmean.fit(X)
```

##### Let's plot our cluster centers decided by the model.

```python
centers = kmean.cluster_centers_
print(centers)
```

##### Comparison of original data with clustered results

Now, let us plot our clustered data along with the original data. So, we will be able to see how our model clustered the data.

```python
new_labels = kmean.labels_
# Plot the identified clusters and compare with the answers
fig, axes = plt.subplots(1, 2, figsize=(12,7))
axes[0].scatter(X[:, 0], X[:, 1], c=y, cmap='gist_rainbow', edgecolor='k', s=150)
axes[1].scatter(X[:, 0], X[:, 1], c=new_labels, cmap='jet', edgecolor='k', s=150)
axes[0].set_xlabel('Sepal length')
axes[0].set_ylabel('Sepal width')
axes[1].set_xlabel('Sepal length')
axes[1].set_ylabel('Sepal width')
axes[0].set_title('Original')
axes[1].set_title('Predicted')
```
